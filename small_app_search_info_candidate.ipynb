```python
"""
    small app search informations of cadidate
    library: tkinter, pyodbc
"""

from tkinter import *
from tkinter import messagebox
import pyodbc

conx2022 = pyodbc.connect('DRIVER={ODBC Driver 13 for SQL Server}; SERVER=NAM-S-LAPTOP; Database=BTLPython; UID=BTLPython; PWD=123456789;')
print(pyodbc.drivers())
cursor = conx2022.cursor()

tinh_thanh = {
    '1': 'TP HÀ NỘI', '2': 'TP HỒ CHÍ MINH', '3': 'TP HẢI PHÒNG', '4': 'TP ĐÀ NẴNG', '5': 'HÀ GIANG',
    '6': 'CAO BẰNG', '7': 'LAI CHÂU', '8': 'LÀO CAI', '9': 'TUYÊN QUANG', '10': 'LẠNG SƠN',
    '11': 'BẮC KẠN', '12': 'THÁI NGUYÊN', '13': 'YÊN BÁI', '14': 'SƠN LA', '15': 'PHÚ THỌ',
    '16': 'VĨNH PHÚC', '17': 'QUẢNG NINH', '18': 'BẮC GIANG', '19': 'BẮC NINH',
    '21': 'HẢI DƯƠNG', '22': 'HƯNG YÊN', '23': 'HÒA BÌNH', '24': 'HÀ NAM',
    '25': 'NAM ĐỊNH', '26': 'THÁI BÌNH', '27': 'NINH BÌNH', '28': 'THANH HÓA',
    '29': 'NGHỆ AN', '30': 'HÀ TĨNH',
    '31': 'QUẢNG BÌNH', '32': 'QUẢNG TRỊ', '33': 'THỪA THIÊN - HUẾ', '34': 'QUẢNG NAM',
    '35': 'QUẢNG NGÃI', '36': 'KON TUM', '37': 'BÌNH ĐỊNH', '38': 'GIA LAI',
    '39': 'PHÚ YÊN', '40': 'ĐẮK LẮK',
    '41': 'KHÁNH HÒA', '42': 'LÂM ĐỒNG', '43': 'BÌNH PHƯỚC', '44': 'BÌNH DƯƠNG',
    '45': 'NINH THUẬN', '46': 'TÂY NINH', '47': 'BÌNH THUẬN', '48': 'ĐỒNG NAI',
    '49': 'LONG AN', '50': 'ĐỒNG THÁP',
    '51': 'AN GIANG', '52': 'BÀ RỊA', '53': 'TIỀN GIANG', '54': 'KIÊN GIANG',
    '55': 'TP CẦN THƠ', '56': 'BẾN TRE', '57': 'VĨNH LONG', '58': 'TRÀ VINH', '59': 'SÓC TRĂNG',
    '60': 'BẠC LIÊU',
    '61': 'CÀ MAU', '62': 'ĐIỆN BIÊN', '63': 'ĐĂK NÔNG', '64': 'HẬU GIANG'
}

column = [
    '', '', 'Toán', 'Ngữ văn', 'Ngoại ngữ', 'Vật lý',
    'Hoá học', 'Sinh học', 'Lịch sử', 'Địa lý', 'GDCD', 'KHTN', 'KHXH', 'TBTN'
]

app = Tk()
app.title('Tra cứu điểm thi')
app.geometry('500x650')
app.resizable(width=False, height=False)
img = PhotoImage(file='kinh_lup.ico')
app.iconphoto(False, img)
text = StringVar()
lb = [0] * 15
phan_cach = None

def process():
    global phan_cach
    sbd = text.get()
    if len(sbd) != 8:
        messagebox.showerror('Thông báo!', 'Bạn phải nhập đủ 8 chữ số')
        text.set('')
        return
    
    if sbd[0] == '0':
        sbd = sbd[1:8]
    
    sbd = int(sbd)
    cursor.execute(f'\
        SELECT * FROM THPTQG2022\
        WHERE SBD = {sbd}'
    )
    data = cursor.fetchall()
    if len(data) == 0:
        messagebox.showerror('Thông báo!', 'Không tìm thấy thông tin của thí sinh')
        text.set('')
        return
    
    data = list(data[0])
    sbd = str(sbd)
    KHTN = (data[4] + data[5] + data[6]) / 3
    KHXH = (data[7] + data[8] + data[9]) / 3
    TBTN = (max(KHTN, KHXH) + data[1] + data[2] + data[3]) / 4
    data = data + [KHTN, KHXH, TBTN]
    phan_cach = Label(
        app,
        text='---------------------*******---------------------',
        font=('Times New Roman', 10, 'bold'),
        fg='black'
    )
    phan_cach.place(x = 0, y = 310)
    
    lb[0] = Label(
        app,
        text='Sở GDĐT ' + (tinh_thanh[sbd[0]] if len(sbd) == 7 else tinh_thanh[sbd[0:2]]),
        font=('Arial', 10, 'bold'),
        fg='black'
    )
    lb[0].place(x = 50, y = 330)
    
    lb[1] = Label(
        app,
        text='Điểm thi của thí sinh mang SBD ' + ('0' + str(sbd) if len(sbd) == 7 else str(sbd)),
        font=('Arial', 10, 'bold'),
        fg='black'
    )
    lb[1].place(x = 50, y = 350)
    
    for i in range(2, 14):
        lb[i] = Label(
            app,
            text=column[i] + ': ' + f'{data[i - 1]:.2f}',
            font=('Arial', 10, 'bold'),
            fg='black'
        )
        lb[i].place(x = 50, y = 350 + 20 * (i - 1))
    

def process_clear():
    text.set('')
    phan_cach.destroy()
    for i in range(0, 14):
        lb[i].destroy()

lb1 = Label(app, text='Tra cứu điểm thi THPTQG năm 2022',
            font=('Times New Roman', 14, 'bold', 'underline'),
            fg='#d91427',
            bg='#14a4d9').place(x = 95, y = 100)

lb2 = Label(app, text='Nhập SBD (đầy đủ 8 chữ số)',
            font=('Arial', 14, 'bold'),
            fg='black',
            bg='white').place(x = 120, y = 150)

textbox = Entry(app, textvariable=text, width=30, font=('Consolas', 12)).place(x = 115, y = 200)
lookup = Button(app, text='Search', font=('Hack', 10, 'bold', 'italic'), fg='#5a6982', command=process).place(x = 250, y = 250)
clear = Button(app, text='Clear', font=('Hack', 10, 'bold', 'italic'), fg='#5a6982', command=process_clear).place(x = 320, y = 250)

app.mainloop()
```
